**Problem Statement:**

The rapid adoption of AI across industries and countries is transforming the global economy, influencing job markets, revenue streams, consumer trust, and regulatory policies. However, the degree of AI adoption, its economic impact, and the societal responses vary significantly by geography, industry, and regulation status.
This study aims to analyze patterns and correlations in AI adoption rates, content generation volumes, job displacement, revenue changes, and human-AI collaboration across countries and industries. The goal is to identify key drivers of AI adoption, assess trade-offs between job loss and revenue growth, and understand how regulation and consumer trust impact market share of AI companies.

**1. AI Adoption Trends**

* **Industries and countries have the highest AI adoption rates**

```python
# chart showing industries and countries have the highest AI adoption rates

df= mydata.groupby(['Industry', 'Country'])['AI Adoption Rate (%)'].mean().reset_index() 
sns.barplot(data=df, x='AI Adoption Rate (%)', y='Industry', palette='viridis',errorbar=('ci', False))
plt.title('AI Adoption Rates by Industry ')
plt.xlabel('AI Adoption Rate')
plt.ylabel('Industry')
plt.show()

```

```python
# Ai adoption rate by country
df= mydata.groupby(['Country'])['AI Adoption Rate (%)'].mean().sort_values(ascending=False).reset_index() 
sns.barplot(data=df, x='AI Adoption Rate (%)', y='Country', palette='viridis',errorbar=('ci', False))
plt.title('AI Adoption Rates by Country ')
plt.xlabel('AI Adoption Rate')
plt.ylabel('Country')
plt.show()

```

**Insights :**
1. AI adoption is consistently high across the board  no country or industry in the charts drops below ~48%.

2. Certain sectors (Gaming, Finance) and regions (UK, France) are pushing the upper limits, showing they could be early leaders in AI-driven economic advantage.

3. The spread between the highest and lowest values is relatively small, meaning the gap in AI adoption is narrowing globally.

**2. Economic Impact**

* **Impact of AI adoption rate on Revenue**

```python
#Does a higher AI adoption rate lead to greater revenue increase?
new = mydata.groupby('Industry')[['AI Adoption Rate (%)', 'Revenue Increase Due to AI (%)']].mean().round(2).sort_values(by='Revenue Increase Due to AI (%)',ascending=False).reset_index()
new.head(10)
sns.set_style("whitegrid")

new_sorted = new.sort_values(by="Revenue Increase Due to AI (%)", ascending=False)

plt.figure(figsize=(12, 6))
barplot = sns.barplot(
    data=new_sorted,
    y='Revenue Increase Due to AI (%)',
    x='Industry',
    hue='Industry',
    palette='Set2',
    dodge=False,
    errorbar=None
)

for p in barplot.patches:
    height = p.get_height()
    barplot.text(
        p.get_x() + p.get_width()/2,  
        height + 0.5,                 
        f'{height:.1f}%',             
        ha='center', va='bottom'
    )

# Titles and labels
plt.title('Revenue Increase vs AI Adoption Rate by Industry', fontsize=16, weight='bold')
plt.ylabel('Revenue Increase Due to AI (%)', fontsize=12)
plt.xlabel('Industry', fontsize=12)

plt.tight_layout()
plt.show()
```

* **Job Loss due to AI**

```python
#by industry
df = mydata.groupby('Industry')[['Job Loss Due to AI (%)',]].mean().sort_values(ascending=False).reset_index()
df.head()
sns.set_style("whitegrid")
plt.figure(figsize=(12, 5))
barplot = sns.barplot(
    data=df,
    y='Job Loss Due to AI (%)',
    x='Industry',
    palette='Set2',
    dodge=True,
    errorbar=None
)


for p in barplot.patches:
    height = p.get_height()
    barplot.text(
        p.get_x() + p.get_width()/2,  
        height + 0.5,                 
        f'{height:.1f}%',             
        ha='center', va='bottom'
    )   
# Titles and labels
plt.title('Job Loss Due to AI by Industry', fontsize=16, weight='bold')
plt.ylabel('Job Loss Due to AI (%)', fontsize=12)
plt.xlabel('Industry', fontsize=12)

plt.tight_layout()
plt.show()


#by country

sns.set_style("whitegrid")
plt.figure(figsize=(12, 5))
barplot = sns.barplot(
    data=df,
    y='Job Loss Due to AI (%)',
    x='Country',
    palette='Set2',
    dodge=True,
    errorbar=None
)


for p in barplot.patches:
    height = p.get_height()
    barplot.text(
        p.get_x() + p.get_width()/2,  
        height + 0.5,                 
        f'{height:.1f}%',             
        ha='center', va='bottom'
    )   
# Titles and labels
plt.title('Job Loss Due to AI by Country', fontsize=16, weight='bold')
plt.ylabel('Job Loss Due to AI (%)', fontsize=12)
plt.xlabel('Countries', fontsize=12)

plt.tight_layout()
plt.show()
```

**Insights :**
1. Industries like Automotive, Media, and Manufacturing show the highest revenue gains from AI adoption, but Manufacturing also experiences the highest job losses, indicating significant automation impact. 
2. Across countries, Canada and Germany report the highest job loss percentages, while the USA shows relatively low job losses despite its AI growthsuggesting differing adaptation strategies. 
3. This highlights that while AI drives revenue, its employment effects vary greatly across industries and nations.

**3. Human-AI Collaboration**

* **Sectors showing the highest collaboration rates between humans and AI**

```python
#Industries showing the highest collaboration rates between humans and AI

df = mydata.groupby(['Industry'])['Human-AI Collaboration Rate (%)'].mean().sort_values(ascending=False).reset_index()
df.head(10)
sns.set_style("darkgrid")
plt.figure(figsize=(12, 5))
barplot = sns.barplot(
    data=df,
    y='Human-AI Collaboration Rate (%)',
    x='Industry',
    palette='Set2',
    dodge=True,
    errorbar=None
)   
for p in barplot.patches:
    height = p.get_height()
    barplot.text(
        p.get_x() + p.get_width()/2,  
        height + 0.5,                 
        f'{height:.1f}%',             
        ha='center', va='bottom'
    )   
# Titles and labels
plt.title('Human-AI Collaboration Rate by Industry', fontsize=16, weight='bold')
plt.ylabel('Human-AI Collaboration Rate (%)', fontsize=12)
plt.xlabel('Industry', fontsize=12)

plt.tight_layout()
plt.show()
```

* **Relationship between human-AI collaboration rate and Consumer trus**

```python
df = mydata.groupby('Industry')[['Human-AI Collaboration Rate (%)','Consumer Trust in AI (%)']].mean().round(2).sort_values(by='Consumer Trust in AI (%)',ascending=True).reset_index()
df.head(10)
sns.set_style("darkgrid")
plt.figure(figsize=(12, 5))
barplot = sns.barplot(
    data=df,
    y='Human-AI Collaboration Rate (%)',
    x='Consumer Trust in AI (%)',
    hue='Industry',
    palette='Set2',
    dodge=True,
    errorbar=None
)   

# Titles and labels
plt.title('Human-AI Collaboration Rate by Consumer trust per Industry', fontsize=16, weight='bold')
plt.ylabel('Human-AI Collaboration Rate (%)', fontsize=12)
plt.xlabel('Consumer Trust', fontsize=12)
plt.legend(title='Industry', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Insights :**
1. Highest Collaboration Rate  Education leads with the highest Human-AI collaboration rate (58.9%), suggesting strong integration of AI tools with human roles in learning environments.

2. Lowest Collaboration Rate  Media shows the lowest collaboration rate (50.0%), potentially indicating slower AI adoption or limited AI-human interaction in this sector.

3. Consumer Trust Variation  Industries with higher collaboration rates dont always have the highest consumer trust. For example:

* Education has the highest collaboration rate but one of the lowest consumer trust scores (~51.65).

* Retail has moderate collaboration (55.4%) but highest consumer trust (~62.73).

4. Inverse Patterns  In some cases, higher collaboration rates correlate with lower consumer trust (e.g., Education, Manufacturing), which may imply concerns about AI replacing human elements.

5. Balanced Sectors  Marketing and Healthcare maintain a balance between collaboration rate (~55%) and trust (~5860%), suggesting healthy coexistence of AI and human roles.

6. Potential Risk Zones  Industries like Education and Legal may need to address trust gaps despite high or moderate AI collaboration.

7. Strong Trust Leaders  Retail and Automotive score highest in consumer trust (~6261) even with moderate AI collaboration, showing effective trust-building strategies.

**4. Trade-offs & Risks**

* **Relationship between job loss due to AI and revenue increase**

```python
plt.figure(figsize=(10, 6))
sns.regplot(
    data=mydata,
    x='Job Loss Due to AI (%)',
    y='Revenue Increase Due to AI (%)',
    scatter_kws={'s': 80, 'alpha': 0.7},
    line_kws={'color': 'red'}
)
plt.title('Job Loss vs Revenue Increase due to AI', fontsize=16, weight='bold')
plt.xlabel('Job Loss Due to AI (%)', fontsize=12)
plt.ylabel('Revenue Increase Due to AI (%)', fontsize=12)
plt.show()



plt.figure(figsize=(10, 6))
sns.regplot(
    data=mydata,
    x='Job Loss Due to AI (%)',
    y='Consumer Trust in AI (%)',
    scatter_kws={'s': 80, 'alpha': 0.7},
    line_kws={'color': 'red'}
)
plt.title('Job Loss vs Consumer Trust in AI', fontsize=16, weight='bold')
plt.xlabel('Job Loss Due to AI (%)', fontsize=12)
plt.ylabel('Consumer Trust in AI (%)', fontsize=12)
plt.show()
```

**Insights :**
1. Slight Positive Link to Revenue  Industries seeing higher job loss due to AI tend to show a small but positive correlation with increased revenue, suggesting some economic benefits despite workforce cuts.

2. Trust Not Severely Damaged  Consumer trust in AI also shows a slight upward trend with job losses, meaning job reductions havent strongly eroded public trust overall.

3. Weak Correlation Overall  In both cases, the relationships are weak, indicating other factors (beyond job loss) heavily influence revenue growth and trust levels.

4. High Variability  At every level of job loss, theres a wide spread in both revenue gains and trust scores, showing the impact of AI adoption varies greatly by context.

**5. Market Trends of AI in last 5 Years**

* **Global Trend in Market Shares of AI Companies**

```python
market_trend = mydata.groupby('Year', as_index=False)['Market Share of AI Companies (%)'].mean()

sns.lineplot(
    data=market_trend,
    x='Year',
    y='Market Share of AI Companies (%)',
    marker='o',
    color='blue'
)
plt.title('Global Trend in Market Share of AI Companies', fontsize=16, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Market Share (%)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

**Insights :**
1. Steady Early Growth  From 2020 to 2021, AI companies global market share rose sharply from ~24.5% to ~27.3%.

2. Minor Dip in 2022  There was a small decline in 2022, suggesting short-term challenges or market corrections.

3. Strong Peak in 2024  Market share climbed steadily after 2022, hitting its highest point (~29.6%) in 2024.

4. Sharp Decline in 2025  The drop to ~25.7% in 2025 signals potential saturation, increased competition, or reduced adoption.

* **Trend in top Tech Countries**

```python
# Filter for selected countries
selected_countries = ['India', 'USA', 'UK', 'China', 'Japan']
countries_data = mydata[mydata['Country'].isin(selected_countries)]

# Plot
plt.figure(figsize=(14,6))
sns.lineplot(
    data=countries_data,
    x='Year',
    y='Market Share of AI Companies (%)',
    hue='Country',
    marker='o',
    linewidth=2
)

plt.title('Trend in Market Share of AI Companies (Selected Countries)', fontsize=16, weight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Market Share (%)', fontsize=12)
plt.ylim(0, 100)
plt.grid(True, alpha=0.3)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Insights :**

1.  **USA's Dominance and Subsequent Decline:** The USA consistently held the highest market share in the early years, peaking around 2021. However, its market share experienced a significant decline from that point, dropping below 30% by 2025 and falling behind other countries like the UK.

2.  **UK's Surge to the Top:** The UK showed a notable and dramatic increase in market share, especially between 2023 and 2024. This surge propelled the UK to become the leading country in AI market share by 2024, surpassing all other nations shown on the chart. While it experienced a slight dip in 2025, it still remained a strong competitor.

3.  **China's Volatile but Upward Trajectory:** China's market share was relatively low at the beginning of the period but showed an overall upward trend. It experienced a peak in 2023, followed by a slight dip, and then recovered to finish with a higher market share in 2025 than it started with in 2020. This indicates a growing presence in the AI market, albeit with some volatility.

4.  **India's Consistent Growth:** India demonstrated a steady and relatively consistent increase in its AI market share throughout the entire period. Starting with one of the lowest shares in 2020, it showed a gradual but continuous rise, indicating a stable and expanding presence in the global AI landscape by 2025.
